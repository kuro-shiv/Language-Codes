```python
categories=['A','B','C','D','E']
values=[10,24,36,18,30]
```

```python
bars=plt.bar(categories,values,color='orange',width=0.5)
plt.title('Orange bar with categories')
plt.xlabel=('Categories')
plt.ylabel=('Values')
for bar in bars:
    height=bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2,height + 0.5,
            str(height),
            ha='center',va='bottom',fontsize=10,color='black')
    plt.show()
```

```python
import matplotlib.pyplot as plt

categories = ['A', 'B', 'C', 'D', 'E']
values = [10, 24, 38, 16, 30]

# Create a horizontal bar chart
bars = plt.barh(categories, values, color='skyblue')

# Adding titles and labels
plt.title('Bar chart with value labels')
plt.xlabel('Values')          # X-axis shows values
plt.ylabel('Categories')      # Y-axis shows categories

# Adding value labels on top of each bar
for bar in bars:
    plt.text(bar.get_width() + 0.5,
             bar.get_y() + bar.get_height() / 2,
             str(bar.get_width()),
             va='center', fontsize=10, color='black')

plt.show()
```

```python
years = [2018, 2019, 2020, 2021, 2022, 2023, 2024]
pass_percentage = [83.01, 91.10, 88.78, 99.04, 92.71, 93.12, 94.56]

plt.bar(years, pass_percentage, color='green')
plt.title('CBSE Pass Percentage (2018-2024)')
plt.xlabel('Year')
plt.ylabel('Pass Percentage')   

for i in range(len(years)):
    plt.text(years[i], pass_percentage[i] + 0.5, f'{pass_percentage[i]}%', ha = 'center', fontsize = 9)

plt.grid(axis = 'y', linestyle = '--', alpha=0.7)
plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Define the data correctly
data = {
    'Class': ['Class 6', 'Class 7', 'Class 8', 'Class 9', 'Class 10'],
    '2022': [120, 115, 130, 125, 110],
    '2023': [125, 110, 135, 130, 112],
    '2024': [128, 120, 140, 135, 115]
}

df = pd.DataFrame(data)

x = np.arange(len(df['Class']))
bar_width = 0.25

# Plot bars
plt.bar(x - bar_width, df['2022'], width=bar_width, label='2022', color='skyblue')
plt.bar(x, df['2023'], width=bar_width, label='2023', color='orange')
plt.bar(x + bar_width, df['2024'], width=bar_width, label='2024', color='green')

# Add title and labels
plt.title('Student Strength Analysis (2022–2024)')
plt.xlabel('Class')  # ✅ Don't overwrite plt.xlabel
plt.ylabel('Number of Students')
plt.xticks(x, df['Class'])
plt.legend()

# Add value labels on top of bars
for i in range(len(x)):
    plt.text(x[i] - bar_width, df['2022'][i] + 1, str(df['2022'][i]), ha='center', fontsize=8)
    plt.text(x[i], df['2023'][i] + 1, str(df['2023'][i]), ha='center', fontsize=8)
    plt.text(x[i] + bar_width, df['2024'][i] + 1, str(df['2024'][i]), ha='center', fontsize=8)

plt.grid(axis='y', linestyle='--', alpha=0.6)
plt.tight_layout()
plt.show()
```
