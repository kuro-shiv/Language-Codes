```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data=[23,45,56,67,45,23,23,45,67,89,90,34,23,45,56,23,67]
plt.hist(data,bins=5,color='orange',edgecolor='black')
plt.xlabel('Marks Range')
plt.ylabel('Number of Students')
plt.title('Marks distribution Histogram')
plt.show()
```

```python
data=[
    [45,67,78,56,89],
    [50,70,65,60,75],
    [75,68,72,58,80],
    [96,68,76,58,90],
    [85,68,72,78,75]
]
df=pd.DataFrame(data,columns=['Maths','English','Science','History','Computer'])
print(df)
```
